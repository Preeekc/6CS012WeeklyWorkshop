```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Load dataset
df=pd.read_csv("SampleSuperstore.csv")

# Setting style
sns.set(style="whitegrid")

# Matplotlib & Seaborn Visualizations
fig, axes = plt.subplots(3, 4, figsize=(20, 15))

# Boxplots
sns.boxplot(x='Category', y='Sales', data=df, ax=axes[0, 0]).set_title("Boxplot - Sales by Category")
sns.boxplot(x='Category', y='Profit', data=df, ax=axes[0, 1]).set_title("Boxplot - Profit by Category")
sns.boxplot(x='Region', y='Sales', data=df, ax=axes[0, 2]).set_title("Boxplot - Sales by Region")
sns.violinplot(x='Region', y='Profit', data=df, ax=axes[0, 3]).set_title("Violin Plot - Profit by Region")

# Histograms
sns.histplot(df['Sales'], bins=20, kde=True, ax=axes[1, 0]).set_title("Histogram - Sales Distribution")
sns.histplot(df['Profit'], bins=20, kde=True, ax=axes[1, 1]).set_title("Histogram - Profit Distribution")
axes[1, 2].hist(df['Discount'], bins=15, color='skyblue', edgecolor='black')
axes[1, 2].set_title("Histogram - Discount Distribution")
sns.kdeplot(df['Quantity'], shade=True, ax=axes[1, 3]).set_title("Density Plot - Quantity")

# Scatterplots
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df, ax=axes[2, 0]).set_title("Scatterplot - Sales vs Profit")
axes[2, 1].scatter(df['Sales'], df['Discount'], color='red', alpha=0.5)
axes[2, 1].set_title("Scatterplot - Sales vs Discount")
sns.regplot(x='Sales', y='Profit', data=df, ax=axes[2, 2]).set_title("Regression Plot - Sales vs Profit")
sns.lineplot(x=df.index, y='Sales', data=df, ax=axes[2, 3]).set_title("Line Graph - Sales Over Time")

plt.tight_layout()
plt.show()

# Plotly Express Visualizations
fig1 = px.box(df, x='Category', y='Sales', title="Plotly Boxplot - Sales by Category")
fig2 = px.histogram(df, x='Sales', nbins=20, title="Plotly Histogram - Sales Distribution")
fig3 = px.scatter(df, x='Sales', y='Profit', color='Category', title="Plotly Scatterplot - Sales vs Profit")
fig4 = px.line(df, x=df.index, y='Sales', title="Plotly Line Graph - Sales Over Time")

# Display Plotly graphs
fig1.show()
fig2.show()
fig3.show()
fig4.show()
```
